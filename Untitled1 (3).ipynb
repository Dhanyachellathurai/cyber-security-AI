```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 14))   # Large dashboard size

# -----------------------------------------
# 1. Scatterplot (Top-left)
# -----------------------------------------
plt.subplot(2, 2, 1)
sns.scatterplot(
    data=df,
    x=' Flow Duration',
    y=' Total Fwd Packets',
    s=8,
    alpha=0.4
)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Flow Duration (log scale)")
plt.ylabel("Total Forward Packets (log scale)")
plt.title("Flow Duration vs Total Fwd Packets", fontsize=12)

# -----------------------------------------
# 2. Top 20 Destination Ports (Top-right)
# -----------------------------------------
plt.subplot(2, 2, 2)
df[' Destination Port'].value_counts().head(20).plot(kind='bar')
plt.title("Top 20 Destination Ports", fontsize=12)
plt.xlabel("Destination Port")
plt.ylabel("Count")

# -----------------------------------------
# 3. Histogram of Flow Duration (Bottom-left)
# -----------------------------------------
plt.subplot(2, 2, 3)
plt.hist(df[' Flow Duration'], bins=50)
plt.title("Histogram of Flow Duration", fontsize=12)
plt.xlabel("Flow Duration")
plt.ylabel("Frequency")

# -----------------------------------------
# 4. Pie Chart of Labels (Bottom-right)
# -----------------------------------------
plt.subplot(2, 2, 4)
counts = df[' Label'].value_counts()
plt.pie(counts, labels=None, startangle=90)
plt.title("Label Distribution", fontsize=12)
plt.legend(counts.index, loc="upper left", fontsize=10)

# Final layout fix
plt.tight_layout()
plt.savefig("Report5.png")
plt.show()
```
