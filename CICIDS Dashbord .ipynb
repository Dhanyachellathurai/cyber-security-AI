```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df[' Flow Duration'],df[' Total Fwd Packets'],color="red")
plt.title("'Flow Duration vs  Total Fwd Packets" )
plt.xlabel("'Flow Duration")
plt.ylabel("Total Fwd Packets")
plt.show()
```

```python
top_20_ports = df[' Destination Port'].value_counts().nlargest(20)

plt.figure(figsize=(12, 6))
plt.bar(top_20_ports.index.astype(str), top_20_ports.values, color="red")
plt.title("Top 20 Most Used Destination Ports")
plt.xlabel("Destination Port")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.hist(df[' Flow Duration'], bins=50)
plt.title("Histogram of Flow Duration")
plt.xlabel("Flow Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
counts = df[' Label'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", pctdistance=0.80)

circle = plt.Circle((0,0), 0.70, color='white')
plt.gca().add_artist(circle)

plt.legend(
    [f"{counts.index[i]} = {counts.iloc[i]}" for i in range(len(counts))],
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=9
)

plt.title("Label Distribution")
plt.tight_layout()
plt.show()
```
